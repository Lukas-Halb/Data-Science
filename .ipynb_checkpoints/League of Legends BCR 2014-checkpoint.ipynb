```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.options.mode.chained_assignment = None
import bar_chart_race as bcr
```

```python
df = pd.read_csv('2014_LoL_esports_match_data_from_OraclesElixir_20200824.csv')
```

```python
df.head()
```

```python
championcounttotal2014 = pd.DataFrame(columns=['Champ', 'Played'])

for champ in df['champion'].unique():
    champcount = len(df[df['champion'] == champ])
    row = pd.Series({'Champ': champ, 'Played': champcount})
    championcounttotal2014 = championcounttotal2014.append(row, ignore_index=True)
    
    
championcounttotal2014.head()
```

```python
championsplayedbcr = df[['date','champion']]
for i in championsplayedbcr:
    championsplayedbcr['values'] = 1
championsplayedbcr['date2'] = championsplayedbcr['date'].str.split(' ').str[0]
championsplayedbcr = championsplayedbcr.drop(columns='date')
```

```python
championbrc = pd.DataFrame(columns=['date','Champ', 'Played'])

for date in championsplayedbcr['date2'].unique():
    for champ in championsplayedbcr['champion'].unique():
        Played = championsplayedbcr.loc[(championsplayedbcr['date2'] == date) & (championsplayedbcr['champion'] == champ), 'values'].sum()
        row = pd.Series({'date': date, 'Champ': champ, 'Played' : Played})
        championbrc = championbrc.append(row, ignore_index=True)        
```

```python
championbrc["Played"] = championbrc["Played"].astype(str).astype(int)
```

```python
championbrc = championbrc.pivot_table(index='date', columns='Champ', values='Played')
```

```python
championbrc.head()
```

```python
championbrc = championbrc.fillna(0)
championbrc = championbrc.cumsum()
```
